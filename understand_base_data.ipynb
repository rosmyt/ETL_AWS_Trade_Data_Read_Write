```python
s3 = boto3.resource('s3')
bucket = s3.Bucket('xetra-1234')
```

```python
bucket_obj = bucket.objects.filter(Prefix='2022-03-15')
objects = [obj for obj in bucket_obj]
```

```python
csv_obj = bucket.Object(key='2022-03-15/2022-03-15_BINS_XETR10.csv').get().get('Body').read().decode('utf-8')
```
